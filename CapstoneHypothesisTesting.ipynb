# 1.  CHI SQUARE TEST of INDEPENDENCE

Gender and Product Category 1

NULL HYPOTHESIS = There is no significant relation between gender and product category  

ALTERNATE HYPOTHESIS = There is a significant relation between gender and product category 

```python
##creating contingency table
GP_cont_table = pd.crosstab(df3['Gender'], df3['Product_Category_1'])
GenderProductCategory  = pd.DataFrame(GP_cont_table)
GenderProductCategory
```

```python
chi2, pval, dof, expected = chi2_contingency(GP_cont_table)
```

```python
chi2
```

```python
pval
```

```python
dof
```

```python
expected
```

```python
print("CONCLUSION:")

significance_level = 0.05

if pval < significance_level:
    print("Reject NULL Hypothesis: There is significant relation between the Gender and Product_Category_1")
else:
    print("Accept Null Hypothesis: There is NO significant relation between Gender and Product_Category_1")
```
