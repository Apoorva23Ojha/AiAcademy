# 1.  CHI SQUARE TEST OF INDEPENDENCE

## Gender and Product Category 1

NULL HYPOTHESIS = There is no significant relation between gender and product category  

ALTERNATE HYPOTHESIS = There is a significant relation between gender and product category 

```python
#creating contingency table for gender and product category 

GP_cont_table = pd.crosstab(df3['Gender'], df3['Product_Category_1'])
GenderProductCategory  = pd.DataFrame(GP_cont_table)
GenderProductCategory
```

```python
#Performing Chi square Test

chi2, pval, dof, expected = chi2_contingency(GP_cont_table)
```

```python
print("Chi Square Statistics = ", chi2)
```

```python
print("P-value = ",pval)
```

```python
print("Degree of Freedom = ", dof)
```

```python
print("Expected Frequencies = \n")
print(expected)
```

```python
print("CONCLUSION:\n")

significance_level = 0.05

if pval < significance_level:
    print("Reject NULL Hypothesis: There is significant relation between the Gender and Product_Category_1")
else:
    print("Accept Null Hypothesis: There is NO significant relation between Gender and Product_Category_1")
```

# 2. Z- Test

## Taking Sample Population as Male Customers:

NULL HYPOTHESIS:  Mean Purchase Amount of Male Customers = Mean Purchase Amount of all Customer (Sample is representative of the population)

ALTERNATE HYPOTHESIS : Mean Purchase Amount of Male Customers > Mean Purchase Amount of all Customer (Sample is not representative of the population)

Therefore we perform : 1 Tail Test (Right Tail)

```python
df3.describe()
```

```python
print("Population Count : ")
np = df3['Purchase'].count()
np
```

```python
print("Mean Purchase Amount for whole Population : ")
popMean = df3['Purchase'].mean()
popMean
```

```python
print("Sample Population : Purchase made by Male Customers")
mdf = pd.DataFrame(df3[df3['Gender']== 'M'])
mdf
```

```python
print("Sample Population Count = ")
ns = mdf['Purchase'].count()
ns
```

```python
print("Mean Purchase Amount of Sample Population")

sampleMean = mdf['Purchase'].mean()
sampleMean
```

```python
#performing z-test
zstats , pval = test.ztest(mdf['Purchase'],value=popMean,alternative='larger')
zstats
```

```python
print("P-Value = ", pval)
```

```python
significant_level  =0.05
print("CONCLUSION: \n")

if pval > significant_level:
    print("Accept the NULL Hypothesis: Mean Purchase Amount for Males customers is equal to that of all customers. \nThus we can say for purchase amount Male population is representative of all cutomers poplulation ")
else: 
    print("Reject the NULL Hypothesis: Mean Purchase Amount for Male customers is greater than that of all customers.")
```

# ANOVA Testing

## City_Category and Purchase

NULL HYPOTHESIS : Mean Purchase Amount for each City Category is equal (no variation in means of groups). 

ALTERNATE HYPOTHESIS : At least, one City Category's Mean Purchase Amount is different from other categories.

```python
Adf = pd.DataFrame(df3[df3['City_Category']=='A']['Purchase'])
Adf
```

```python
Bdf = pd.DataFrame(df3[df3['City_Category']=='B']['Purchase'])
Bdf
```

```python
Cdf = pd.DataFrame(df3[df3['City_Category']=='C']['Purchase'])
Cdf
```

```python
F, p = st.f_oneway(Adf, Bdf, Cdf)
```

```python
print("F-Statistics = ", F)
```

```python
print("P-value for significance is: ", p)
```

```python
print("CONCLUSION: \n")

significant_level  =0.05


if p < significant_level:
    print("Reject Null Hypothesis: At least, one City Category's Mean Purchase Amount is different from other categories ")
else:
    print("Accept Null Hypothesis: Mean Purchase Amount for each City Category is equal (no variation in means of groups) ")
```
